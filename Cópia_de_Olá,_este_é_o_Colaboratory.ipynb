# Importação das Planilhas


```
import pandas as pd



tabela_fato = pd.read_csv('stark_fato.csv',index_col=0, encoding='latin1', sep=";",decimal = ',', thousands='.')
tabela_meta = pd.read_csv('stark_meta_prod.csv',index_col=0, encoding='latin1', sep=";", decimal = ',', thousands='.')

display(tabela_fato)
display(tabela_meta)
```

# Editando o cabeçalho

```
tabela_fato=tabela_fato.rename(columns={"cod_sku":"codigo cliente","cod_cliente":"local da venda", "timestamp":"Data", "yhat":"Quantidade"})
tabela_meta=tabela_meta.rename(columns={"cod_sku":"codigo cliente", "prod_agg1":"Descrição", "prod_agg2":"Grupo"})

```

# Juntando as tabelas

```
Planilha_Consolidada = tabela_fato.merge(tabela_meta, on="codigo cliente")
Planilha_Consolidada
```

# Correção do Texto

```
Planilha_Consolidada['Descrição'] = Planilha_Consolidada['Descrição'].replace('Ã§Ã£o', 'ção', regex=True)
```

# Convertendo a coluna "data da venda" para datetime

```
Planilha_Consolidada.Data=pd.to_datetime(Planilha_Consolidada.Data)
```

```
Planilha_Consolidada.Data.dt.to_period('M')
```

```
Planilha_Consolidada['Data'] = Planilha_Consolidada.Data.dt.to_period('M').astype(str)
```

# Vendas por Mês

```
Vendas_Mes = Planilha_Consolidada.groupby(by='Data').Quantidade.sum()
```

# Calculando o valor Máximo das vendas por Mês


```
Max_de_Vendas = Vendas_Mes.values.max()
```
